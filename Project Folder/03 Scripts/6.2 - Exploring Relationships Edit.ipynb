# 06 - Creating Heatmap Matrix

```python
df_playoffsbio.corr()
```

```python
plt.show()
```

```python
# creating a subset for matrix
df_submatrix = df_playoffsbio[['playerId','season','games_played','icetime','year','month','day','weight','height','primaryNumber']]
```

```python
f = plt.figure(figsize=(12, 12)) # figure size
plt.matshow(df_submatrix.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_submatrix.shape[1]), df_submatrix.columns, fontsize=10, rotation=45) # x axis labels
plt.yticks(range(df_submatrix.shape[1]), df_submatrix.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

the above correlation matrix shows that the strongest correlation is between playerid and year, followed by playerid and season, season and year and height and weight. There is a slight negative correlation between year and weight that would be interesting to delve deeper into. The other relationships are not strong.

```python
# save figure above
f.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'correlation_matrix.png'))
```

# 10 - Categorical Plot

```python
# creating a categorical plot that splits weight into categories and then plots it against ice time.
sns.histplot(df_submatrix['weight'], bins = 20, kde = True)
```

dividing players into three weight categories. Lightweight (< 180), Middleweight (180 > 220), Heavyweight (220 <)

```python
# labeling the weights
df_playoffsbio.loc[df_playoffsbio['weight'] < 180, 'Weight category'] = 'Lightweight'
df_playoffsbio.loc[(df_playoffsbio['weight'] >= 180) & (df_playoffsbio['weight'] < 220), 'Weight category'] = 'Middleweight'
df_playoffsbio.loc[df_playoffsbio['weight'] >= 220, 'Weight category'] = 'Heavyweight'
```
